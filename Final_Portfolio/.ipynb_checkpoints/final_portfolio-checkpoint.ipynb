### Investigation 1 - Trend of Bestsellers Genre Over The Years

For the first investigation, we shall check the trend of the bestsellers genre over the years. Are fictions or non-fictions more popular each year? Is there any change in trends over the years?

For this investigation, we will count the number of fiction and non-fiction bestsellers each year, and plot a line graph to show the trend.

First, we count the number of bestsellers for each genre and year, and save to a new dataframe.

```python
genre_distribution = bestsellers.groupby(["Year", "Genre"])["Name"].count().reset_index()
```

```python
genre_distribution
```

After that, we plot a line graph to show the trend.

```python
# set figure size and background
plt.figure(figsize=(15,10))
sns.set_style("ticks")

# line plotting
ax = sns.lineplot(data=genre_distribution, x="Year", y="Name", hue="Genre")

# display each year on x-axis and size of numbers
ax.tick_params(labelsize=14)
ax.set_xticks(range(2009, 2020))
ax.set_xticklabels(range(2009, 2020))

# add title and labels
plt.title("Number of Fiction and Non Fiction Bestsellers from 2009 to 2019", fontsize=18)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.legend(fontsize=14)

plt.show()
```

Based on the graph above, we discovered a few things:
1. For most of the year, more bestsellers are non-fiction than fiction.
2. There were two times when there was a drop of non-fiction bestsellers, which were 2012-2014 and 2015-2017.
3. In 2014, the number of fiction bestsellers surpassed those of non-fiction.
4. In 2015, the number of non-fiction bestsellers were significantly higher than that of fiction bestsellers. It was the year with the most difference.

It would be interesting to check what was the bestsellers in 2014 and 2015.

### Investigation 2 - Bestsellers that Lasts a Long Period

For the second investigation, we shall check if there are any very famous books which are bestsellers for multiple years. It would be very impressive if a book is a bestseller for 5 years or more, so let's see if there are such books and what are they.

To obtain these titles, we first group the bestsellers by name and count the number of unique years. Then, we extract all titles with a count of 5 or above. As we would also like to check the other information of the title (e.g. Author, Ratings), we inner merge the extracted bestsellers dataframe with the original one, and cleaning the new dataframe, such as dropping columns not in interest and renaming columns.

```python
year_counts = bestsellers.groupby("Name")["Year"].count().reset_index()
year_counts = year_counts[year_counts["Year"] >= 5].rename(columns={"Year": "Years in Bestsellers"})
bestsellers_over_5_years = pd.merge(bestsellers, year_counts, left_on="Name", right_on="Name").drop_duplicates(subset="Name").sort_values(by=["Years in Bestsellers", "Genre"], ascending=False).reset_index().drop(columns=["index", "Year"])
```

```python
bestsellers_over_5_years
```

There are a total of 17 books that were bestsellers for 5 years or above. Among them, there are 12 non-fiction and only 5 fiction. Most of the 17 books have a rating of over 4.5, except StrengthsFinder2.0 with a rating of 4.0 and The Official SAT Study Guide with a rating of 4.4. The bestseller with the most years in bestsellers is the Publication Manual of the American Psychological Association, which is a bestseller for 10 years. It is followed by StrengthsFinder 2.0 with 9 years and Oh, the Places You'll Go! with 8 years.

### Investigation 3 - Authors with Multiple Bestsellers

For the third investigation, we shall check if any authors wrote multiple books that got into bestsellers. Let's check if any authors have 5 different titles that got into bestsellers.

First, we remove rows with duplicated book names. Then, we count the number of titles for each author. Further, we filter the authors who have at least 5 individual bestsellers. At last, we clean the dataframe, such as renaming columns and arrange the authors by the number of bestsellers in descending order.

```python
authors_over_5_books = bestsellers.drop_duplicates(subset="Name").groupby("Author")["Name"].count().reset_index()
authors_over_5_books = authors_over_5_books[authors_over_5_books["Name"] >= 5].rename(columns={"Name": "Number of Bestsellers"}).sort_values(by=["Number of Bestsellers"], ascending=False).reset_index(drop=True)
```

```python
authors_over_5_books
```

9 authors have 5 bestsellers or more. Who are they? Let's check each author and their bestsellers.

We create a database with authors and their publications using a Python dictionary. The keys of the dictionary will be the author's names, and the values will be a filtered dataframe with the names of the unique titles, and their respective information. 

As one individual title may be bestsellers for multiple years, we pick the row with the earliest year, which means the first year the title was listed as bestsellers.

```python
sorted_bestsellers = bestsellers.sort_values(by=["Year"], ascending=True)

authors_dict = {}
for author in authors_over_5_books["Author"].values:
    authors_dict[author] = sorted_bestsellers[sorted_bestsellers["Author"] == author].drop_duplicates(subset="Name").reset_index(drop=True)
```

### Investigation 5 - Can we predict the genre of a book based on its name?

Is it possible that judging from the name of the book, we can predict if the book is fiction or non-fiction? We shall find out in this fifth and last session, where we will create a machine learning model and test if we can predict the genre of a book with accuracy.

We will use naive bayes model to predict the genre of the book. First, we import the necessary libraries.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = bestsellers["Name"]
labels = bestsellers["Genre"]
```

```python
data
```

```python
labels
```

```python
vectorizer = CountVectorizer()
vectorizer.fit(data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=17)
```

```python
x_train_count = vectorizer.transform(x_train)
x_test_count = vectorizer.transform(x_test)
```

```python
classifier = MultinomialNB()
classifier.fit(x_train_count, y_train)
```

```python
predictions = classifier.predict(x_test_count)
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions, labels=classifier.classes_))
```
